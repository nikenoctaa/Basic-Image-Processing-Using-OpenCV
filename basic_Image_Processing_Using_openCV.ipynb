```
import numpy as np
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
import cv2 #import open cv library
```

```
img = cv2.imread('plant.jpg')
```

```
type(img)
```

```
plt.imshow(img)
titles = ['image using OpenCV (BGR)']
```

```
fin_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(fin_img)
titles = ['Change color from BGR to RGB']
```

```
#grayscale

gray_img=cv2.imread('plant.jpg', cv2.IMREAD_GRAYSCALE)
```

```
plt.imshow(gray_img,cmap='gray')
titles = ['change color to gray scale']
```

```
#magmascale

plt.imshow(gray_img,cmap='magma')
titles = ['change color to magma scale']
```

```
#resize 

resize_img = cv2.resize(fin_img,(400,600))
plt.imshow(resize_img)
titles = ['resize (x=400px, y=600px)']
```

```
#resize using ratio

x_ratio = 0.5
y_ratio = 0.3
resize_img2 = cv2.resize(fin_img,(0,0),fin_img,x_ratio,y_ratio)
plt.imshow(resize_img2)
titles = ['resize using ratio (x=0.5, y=0.3)']
```

```
#Laplacian Detection

grayimg = cv2.imread('plant.jpg',cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(grayimg, cv2.CV_64F, ksize=3) #gray image
lap = np.uint8(np.absolute(lap))

plt.imshow(lap)
```

```
#Laplacian Detection

grayimg = cv2.imread('plant.jpg',cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F, ksize=3) #image awal
lap = np.uint8(np.absolute(lap))

plt.imshow(lap)
```

```
#SobelX Detection

grayimg = cv2.imread('plant.jpg',cv2.IMREAD_GRAYSCALE)

sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0) #image awal
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
sobelCombined = cv2.bitwise_or(sobelX, sobelY)
plt.imshow(sobelX)
```

```
#SobelY Detection

grayimg = cv2.imread('plant.jpg',cv2.IMREAD_GRAYSCALE)

sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0) #image awal
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
sobelCombined = cv2.bitwise_or(sobelX, sobelY)
plt.imshow(sobelY)
```
